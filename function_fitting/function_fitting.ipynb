```python
from metakan import KAN, create_dataset
import torch
import matplotlib.pyplot as plt


model = KAN(width=[4,4,2,1], grid=5, k=3, seed=1, hidden_dim=8, embedding_dim=4)
f = lambda x: torch.exp((torch.sin(torch.pi*(x[:,[0]]**2+x[:,[1]]**2))+torch.sin(torch.pi*(x[:,[2]]**2+x[:,[3]]**2)))/2)
dataset = create_dataset(f, n_var=4, train_num=3000)

# train the model
losses = model.fit(dataset, opt="LBFGS", steps=20, lr=1,  update_grid=False);

```
